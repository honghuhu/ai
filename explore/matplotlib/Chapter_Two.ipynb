```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np

ft = FontProperties(fname='/usr/share/fonts/truetype/humor-sans/Humor-Sans.ttf')

%matplotlib inline
```

```python
# 2.1 函数 bar() - 用于画柱状图
x = [1,2,3,4,5,6,7,8]
y = [3,1,4,5,8,2,1,9]
plt.bar(x,y,align='center',color='c',tick_label=['q','a','c','e','r','j','b','p'],hatch='/')
plt.xlabel('box no',fontproperties=ft)
plt.ylabel('box weight')
plt.show()
```

```python
plt.barh(x,y,align='center',color='c',tick_label=['q','a','c','e','r','j','b','p'],hatch='/')
plt.xlabel('box weight (kg)',fontproperties=ft)
plt.ylabel('box no',fontproperties=ft)
plt.show()
```

```python
boxWeight = np.random.randint(0,10,100)
x = boxWeight
bins= range(0,11,1)
plt.hist(x,bins=bins,color='g',histtype='bar',rwidth=1,alpha=0.6)
plt.xlabel("box weight (kg)",fontproperties=ft)
plt.ylabel("Sales volume",fontproperties=ft)
plt.show()
```

```python
kinds = 'Cooking Master Boy','Da Yuzhu','Grandma\'s Home','Fish with fish'
colors = ['c','y','b','g']
soldNums = [0.05,0.45,0.15,0.35]
plt.pie(soldNums,labels=kinds,autopct='%3.1f%%',startangle=60,colors=colors)
plt.title('Assortment',fontproperties=ft)
plt.show()
```

```python
barSlices=12
theta=np.linspace(0.0,2*np.pi,barSlices,endpoint=False)
r=30*np.random.rand(barSlices)
plt.polar(theta,r,color='chartreuse',linewidth=2,marker='*',mfc='b',ms=10)
plt.show()
```

```python
a = np.random.randn(100)
b = np.random.randn(100)
plt.scatter(a,b,s=np.power(10*a+20*b,2),
           c=np.random.rand(100),
           cmap=mpl.cm.RdYlBu,
           marker='o')
plt.show()
```

```python
x = np.linspace(0.5,2*np.pi,20)
y = np.random.randn(20)
plt.stem(x,y,linefmt='-.',markerfmt='o',basefmt='-')
plt.show()
```

```python
x = np.random.randn(1000)
plt.boxplot(x)
plt.xticks([1],['Random number generator AlphaRM'],fontproperties=ft)
plt.ylabel('Stochastic Numerical',fontproperties=ft)
plt.title('随机数生成器抗干扰能力的稳定性',fontproperties=ft)
plt.grid(axis='y',ls=':',lw=1,color='gray',alpha=0.4)
plt.show()
```

```python
x = np.linspace(0.1,0.6,6)
y = np.exp(x)
plt.errorbar(x,y,fmt="bo:",yerr=0.2,xerr=0.02)
plt.xlim(0,0.7)
plt.show()
```
